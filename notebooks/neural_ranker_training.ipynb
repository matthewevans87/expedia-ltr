```python
import json
import matplotlib.pyplot as plt

with open("../runs/ranker_v1/metrics.json") as f:
    metrics = json.load(f)
history = metrics["history"]
epochs = range(1, len(history["loss"]) + 1)

fig, ax = plt.subplots()
ax.plot(epochs, history["loss"], marker="o", label="train")
ax.plot(epochs, history["val_loss"], marker="o", label="val")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.set_title("Neural Ranker — Loss")
ax.legend()
plt.tight_layout()
plt.show()
```
