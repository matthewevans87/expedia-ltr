```python
%cd ..
import pandas as pd, yaml
from src.tf_data import make_twotower_dataset_fast

with open("configs/twotower_v1.yaml") as f:
    cfg = yaml.safe_load(f)

df = pd.read_parquet("data/processed/v1/train.parquet").head(100_000)
ds = make_twotower_dataset_fast(df, cfg["data"], batch_size=2048, shuffle=True, seed=42)

counts = {0.0: 0, 1.0: 0, 5.0: 0}
for batch in ds.take(30):  # ~61k examples
    vals = batch["label"].numpy()
    for k in counts:
        counts[k] += int((vals == k).sum())

print(counts)
```
