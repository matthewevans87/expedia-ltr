```python
import json
import matplotlib.pyplot as plt

with open("../runs/twotower_v1/metrics.json") as f:
    metrics = json.load(f)
loss = metrics["loss_curve"]  # list of floats, one per epoch

fig, ax = plt.subplots()
ax.plot(range(1, len(loss) + 1), loss, marker="o")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.set_title("Two-Tower Training Loss")
plt.tight_layout()
plt.show()
```
